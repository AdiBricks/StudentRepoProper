    <script>
        const image = document.getElementById('image');
        let mouseX = 0;
        let mouseY = 0;

        // Add mousemove event listener to track mouse position
        document.addEventListener('mousemove', (e) => {
            const { clientX, clientY } = e;
            const { offsetWidth, offsetHeight } = image;
            const rotateX = (clientY / offsetHeight - 0.5) * 20; // Adjust the rotation sensitivity
            const rotateY = (clientX / offsetWidth - 0.5) * 40; // Adjust the rotation sensitivity
            image.style.transform = `rotateX(${rotateX}deg) rotateY(${rotateY}deg)`;
            mouseX = clientX;
            mouseY = clientY;
        });

        // Add animation to smoothly move the image
        function rotateImage() {
            const { offsetWidth, offsetHeight } = image;
            const rotateX = (mouseY / offsetHeight - 0.5) * 20; // Adjust the rotation sensitivity
            const rotateY = (mouseX / offsetWidth - 0.5) * 40; // Adjust the rotation sensitivity
            image.style.transform = `rotateX(${rotateX}deg) rotateY(${rotateY}deg)`;
            requestAnimationFrame(rotateImage);
        }
        rotateImage();
    </script>
